```python
x=np.arange(0,10,1)
y=np.arange(0,6,1)
z=np.arange(0,3,1)
```

```python
def xyz_grid(x,y,z):
    x_y_z = np.empty([x.size,y.size,z.size,3])
    
    for i in range(x.size):
        for j in range(y.size):
            for k in range(z.size):
                x_y_z[i][j][k][0]=x[i]
                x_y_z[i][j][k][1]=y[j]
                x_y_z[i][j][k][2]=z[k]
                #print(x_y[i][j][k])
    
    return x_y_z
```

```python
def xyz(x,y,z):
    x_y_z = np.empty([x.size,y.size,z.size])
    
    for i in range(x.size):
        for j in range(y.size):
            for k in range(z.size):
                x_y_z[i][j][k]=x[i]
                x_y_z[i][j][k]=y[j]
                x_y_z[i][j][k]=z[k]
                #print(x_y[i][j][k])
    
    return x_y_z

def xy(x,y):
    x_y = np.empty([x.size,2])
    
    for i in range(x.size):
        x_y_z[i][j]=x[i]
    for j in range(y.size):
        x_y_z[i][j]=y[j]
           
    
    return x_y
```
